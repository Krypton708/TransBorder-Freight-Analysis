```python
import pandas as pd
import numpy as np
```

```python
preprocessed_data = pd.read_csv('/content/Trans_data_new.csv')
preprocessed_data.head(6)
```

```python
preprocessed_data.shape
```

```python
preprocessed_data.drop('Unnamed: 0', axis=1, inplace=True)
preprocessed_data.head(6)
```

```python
mexstate_mode_per_country = preprocessed_data.groupby('Country')['Mexican State'].agg(lambda x: x.mode()[0] if not x.isnull().all() else np.nan)
canprov_mode_per_country = preprocessed_data.groupby('Country')['Canadian Province'].agg(lambda x: x.mode()[0] if not x.isnull().all() else np.nan)
```

```python
# Filling missing 'Mexican State' values based on Country
preprocessed_data['Mexican State'] = preprocessed_data.apply(
    lambda row: mexstate_mode_per_country.get(row['Country'], row['Mexican State'])
    if pd.isna(row['Mexican State']) and row['Country'] == 'Mexico' else row['Mexican State'], axis=1
)
```

```python
# Filling missing 'Canadian Province' values based on Country
preprocessed_data['Canadian Province'] = preprocessed_data.apply(
    lambda row: canprov_mode_per_country.get(row['Country'], row['Canadian Province'])
    if pd.isna(row['Canadian Province']) and row['Country'] == 'Canada' else row['Canadian Province'], axis=1
)
```

*   5 - What are the busiest freight routes based on shipment weight?

```python
preprocessed_data["Route"] = preprocessed_data["U.S. State"] + "  " + preprocessed_data["Country"]
top_routes = preprocessed_data.groupby("Route")["Shipping Weight (Kg)"].sum().nlargest(10)

plt.figure(figsize=(12,6))
sns.barplot(y=top_routes.index, x=top_routes.values, palette="Blues_r")
plt.title("Top 10 Busiest Freight Routes by Shipment Weight")
plt.xlabel("Total Shipping Weight (Kg)")
plt.ylabel("Freight Route")
plt.show()
```
